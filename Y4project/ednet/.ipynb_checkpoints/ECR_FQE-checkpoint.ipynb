```python
import numpy as np
import pandas as pd
import json
import os
import re
import glob
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
plt.style.use('ggplot')
```

### ECR PI

```python
with open('init_state.txt', 'r') as file:
    init_state = file.read()
init_state = json.loads(init_state)
features = list(init_state.keys())

with open('mdp_dict.txt', 'r') as file:
    mdp_dict = file.read()
mdp_dict = json.loads(mdp_dict)

ECR = pd.DataFrame(index = mdp_dict.keys(), columns = ['Start State','ECR','ECR_penalised'])
ECR.index.rename('State representation', inplace = True)
for mdp in mdp_dict.keys():
    feat_to_incl = mdp_dict[mdp]
    starting_state = ''
    for feat in feat_to_incl:
        starting_state += init_state[feat]
    policy = pd.read_csv('policies/policy_{}.csv'.format(mdp))
    policy['state'] = policy['state'].astype('str')
    policy.set_index('state', inplace = True)
    pen_policy = pd.read_csv('policies_penalised/policy_{}.csv'.format(mdp))
    pen_policy['state'] = pen_policy['state'].astype('str')
    pen_policy.set_index('state', inplace = True)
    ECR.loc[mdp, 'Start State'] = starting_state
    ECR.loc[mdp, 'ECR'] = policy.loc[starting_state, 'values']
    ECR.loc[mdp, 'ECR_penalised'] = pen_policy.loc[starting_state, 'values']

ECR.to_csv('Results/ECR.csv')
```

### ECR CQL

```python
results_encoder = {
    'lim_q_mean':'CQL_0',
    'lim_q_qr':'CQL_1',
    'lim_q_iqn':'CQL_2',
    'lim_q_qr_do':'CQL_3',
    'lim_q_qr_bndo':'CQL_4',
    'lim_q_qr_bndo_alph0.5':'CQL_5',
    'lim_q_qr_bndo_alph0':'CQL_6',
    'lim_q_qr_bndo_alph5':'CQL_7',
    'bcq_q_qr_bndo':'BCQ_0'
}
```

```python
folder = 'cql'

path = 'd3rlpy_results/{}'.format(folder)
    
runs = os.listdir(path)
tderror_results = pd.DataFrame()
initval_results =  pd.DataFrame()

for run in runs:
    model_name = re.sub(r'_2021.*$',"", run)
    td_error = pd.read_csv('{}/{}/td_error.csv'.format(path,run), names=['epoch', 'itr', 'td_error'])
    init_state_val = pd.read_csv('{}/{}/initial_state_val.csv'.format(path,run), names=['epoch', 'itr', 'init_val'])
    
    td_error = pd.DataFrame(td_error['td_error'].rename(results_encoder[model_name]))
    init_state_val = pd.DataFrame(init_state_val['init_val'].rename(results_encoder[model_name]))
    
    tderror_results = pd.concat((tderror_results, td_error), axis = 1)
    initval_results = pd.concat((initval_results, init_state_val), axis = 1)
    
tderror_results.index.rename('Epoch', inplace = True)
initval_results.index.rename('Epoch', inplace = True)


file_name = 'cql_{}_results'.format(folder)
with pd.ExcelWriter('Results/{}.xlsx'.format(file_name)) as writer: 
    tderror_results.to_excel(writer, 'TD_error')
    initval_results.to_excel(writer, 'Init_State_Val')
    pd.DataFrame(data  = [results_encoder.keys(), results_encoder.values()]).transpose().to_excel(writer, 'Code')
```

```python
fig, axs = plt.subplots(1,2, figsize = (15,8))
matplotlib.rcParams.update({'font.size': 13})

plot_data = initval_results.loc[:9].drop(columns = ['BCQ_0'])
plot_data = plot_data.reindex(plot_data.columns.sort_values(), axis = 1)
plot_data.reset_index(inplace = True)
sns.lineplot(data = plot_data.melt(id_vars= 'Epoch', var_name='run'), y = 'value', x = 'Epoch', hue = 'run', ax=axs[0])
axs[0].set_ylabel('Init State Value')


plot_data = tderror_results.loc[:9].drop(columns = ['BCQ_0'])
plot_data = plot_data.reindex(plot_data.columns.sort_values(), axis = 1)
plot_data.reset_index(inplace = True)
sns.lineplot(data = plot_data.melt(id_vars= 'Epoch', var_name='run'), y = 'value', x = 'Epoch', hue = 'run', ax = axs[1])
axs[1].set_ylabel('TD Error')

plt.savefig('Saved_Figs/CQL_results.pdf')
plt.show()
```

### FQE Results

```python
fqe_encoder = {
    'fqe_mean':'CQL_0',
    'fqe_qr':'CQL_1',
    'fqe_q_iqn':'CQL_2',
    'fqe_qr_do':'CQL_3',
    'fqe_qr_bndo':'CQL_4',
    'fqe_qr_bndo_alph0.5':'CQL_5',
    'fqe_qr_bndo_alph0':'CQL_6',
    'fqe_qr_bndo_alph5':'CQL_7',
    'fqe_pi': 'MDP_aug4687',
    'fqe_bp_deterministic': 'BP',
    'fqe_bcq':'BCQ_0'
}
```

```python
folder = 'fqe'

path = 'd3rlpy_results/{}'.format(folder)
    
runs = os.listdir(path)
initval_results =  pd.DataFrame()
loss_results = pd.DataFrame()

for run in runs:
    model_name = re.sub(r'_2021.*$',"", run)
    init_state_val = pd.read_csv('{}/{}/init_value.csv'.format(path,run), names=['epoch', 'itr', 'init_val'])
    init_state_val = pd.DataFrame(init_state_val['init_val'].rename(fqe_encoder[model_name]))
    initval_results = pd.concat((initval_results, init_state_val), axis = 1)
    
    loss = pd.read_csv('{}/{}/loss.csv'.format(path,run), names=['epoch', 'itr', 'loss'])
    loss = pd.DataFrame(loss['loss'].rename(fqe_encoder[model_name]))
    loss_results = pd.concat((loss_results, loss), axis = 1)
    
    
initval_results.index.rename('Epoch', inplace = True)
loss_results.index.rename('Epoch', inplace = True)

file_name = 'fqe_results'
with pd.ExcelWriter('Results/{}.xlsx'.format(file_name)) as writer: 
    initval_results.to_excel(writer, 'Init_State_Val')
    loss_results.to_excel(writer, 'Loss')
    pd.DataFrame(data  = [fqe_encoder.keys(), fqe_encoder.values()]).transpose().to_excel(writer, 'Code')
```

```python
fig, axs = plt.subplots(1,2, figsize = (15,8))
matplotlib.rcParams.update({'font.size': 13})

plot_data = initval_results.loc[:9]
plot_data = plot_data.reindex(plot_data.columns.sort_values(), axis = 1)
plot_data.reset_index(inplace = True)
sns.lineplot(data = plot_data.melt(id_vars= 'Epoch', var_name='run'), y = 'value', x = 'Epoch', hue = 'run', ax=axs[0])
axs[0].set_ylabel('Init State Value')


plot_data = loss_results.loc[:9]
plot_data = plot_data.reindex(plot_data.columns.sort_values(), axis = 1)
plot_data.reset_index(inplace = True)
sns.lineplot(data = plot_data.melt(id_vars= 'Epoch', var_name='run'), y = 'value', x = 'Epoch', hue = 'run', ax = axs[1])
axs[1].set_ylabel('FQE Loss')

plt.savefig('Saved_Figs/FQE_results.pdf')
plt.show()
```

```python
matplotlib.rcParams.update({'font.size': 12})
plt.figure(figsize = (10,10))
x = initval_results.columns
vals = initval_results.loc[9].values
plt.scatter(x , vals, color = 'b')
for i, model in enumerate(x):
    loss = loss_results.loc[9,model]
    y = [vals[i]-(loss)**0.5, vals[i]+(loss)**0.5]
    plt.plot([model,model],y,'--+r')
plt.title(r'$s_0$ value  under FQE for each algorithm with error range')
plt.ylabel(r'$s_0$ value')
plt.savefig('Saved_Figs/FQE_vals_with_error.pdf')
plt.show()
```
